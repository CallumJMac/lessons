```python
def decode(message_file):
    """
    Args:
        message_file: .txt file containing the pyramid structure encoded message

    Returns:
        decoded_message: decoded message that is string
    """
    # #read the key file
    with open(message_file, 'r') as file:
        message_file = file.readlines()
    

    message = """
      1
    2 3
    4 5 6
    """

    word_key = {}
    # Split the word key into lines
    lines = message_file.split("\n")
    for line in lines:
        # Just in case the text file contains blank lines
        if len(line) != 0:
            # Split the line into number and word
            number, word = line.strip().split(" ")
            # Convert the number from str to int
            number = int(number)
            # Store values in a dictionary
            word_key[number] = word


    # Initialize an empty list to store the message words
    message_words = []

    # Split the pyramid file into lines
    message_list = message.split("\n")
    for l in message_list:
        # Just in case the text file contains blank lines
        if len(l) != 0:
            # Split each line into a list of numbers and get the last one
            last_num = l.split(" ")[-1]
            if last_num:
                last_num = int(last_num)
                # Append the word to list using word_key
                message_words.append(word_key[last_num])

    # Turn list of words into string
    decoded_message = ' '.join(message_words)
    return decoded_message   
```

```python
def create_staircase(nums):
  while len(nums) != 0:
    step = 1
    subsets = []
    if len(nums) >= step:
      print(step)
      print(nums)
      subsets.append(nums[0:step])
      nums = nums[step:]
      step += 1
    else:
      return False

  return subsets
```
